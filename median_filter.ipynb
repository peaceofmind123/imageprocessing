```python
def medianFilter(M, size=3):
    h,k = M.shape
    temp = np.copy(M)
    m = int((size-1)/2)
    for i in range(0, h - m-1):
        for j in range(0, k - m-1):
            arr = []
            for l in range(size):
                for t in range(size):
                    arr.append(M[i+l, j+t])
            
            arr.sort()
            median = arr[math.floor(len(arr)/2)]
            temp[i+m, j+m] = median
    
    return temp
```
