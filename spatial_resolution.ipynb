```python
def zoomOut(M,n):
    h,k = M.shape
    if h/n != math.floor(h/n) or k/n != math.floor(k/n):
        raise ValueError("Invalid size")
    
    f = int(h/n)
    g = int(k/n)
    print(f,g)
    accum = np.zeros((f,g))
    l=0
    t=0
    for i in range(0,h,n):
        t = 0
        for j in range(0,k,n):
            accum[l,t] = M[i,j]
            t+=1
        
        l+=1
    
    return accum
```

```python
def zoomIn(M, n):
    h, k = M.shape
    accum = np.zeros((n * h, n * k))
    counter = 0
    # row replication
    for i in range(h):
        counter = 0

        for j in range(k):

            for l in range(counter, counter + n):
                accum[i, j + l] = M[i, j]

            counter += n-1

    # column replication
    temp = np.copy(accum)

    for j in range(n * k):
        counter = 0
        for i in range(h):

            for l in range(counter, counter + n):
                accum[i + l, j] = temp[i, j]

            counter += n-1

    return accum
```
