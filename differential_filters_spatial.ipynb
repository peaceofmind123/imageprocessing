```python
import math
import numpy as np
import scipy.signal as ss
import matplotlib.pyplot as plt
```

```python
laplacian = np.array([[0,1,0],[1,-4,1],[0,1,0]])
```

```python
moon = plt.imread('./imgs/moon.tif')
plt.imshow(moon, cmap='gray')
```

```python
fil = ss.correlate2d(moon, laplacian, mode='same')
plt.imshow(fil, cmap='gray')
```

```python
boosted = 1.2*moon - fil
plt.imshow(boosted, cmap='gray')
```

```python
Px = np.array([[-1,-1,-1],[0,0,0],[1,1,1]])
Py = Px.T
```

```python
PfilX = ss.correlate2d(moon, Px, mode='same')
PfilY = ss.correlate2d(moon, Py, mode='same')
```

```python
plt.imshow(PfilX, cmap='gray')
```

```python
plt.imshow(PfilY, cmap='gray')
```

```python
Pfil = (PfilX + PfilY)
```

```python
plt.imshow( Pfil, cmap='gray')
```

```python
Sx = np.array([[-1,-2,-1],[0,0,0],[1,2,1]])
Sy = Sx.T
```

```python
camM = plt.imread('./imgs/cameraMan.tif')
plt.imshow(camM, cmap='gray')
```

```python
SfilX = ss.correlate2d(camM, Sx, mode='same')
SfilY = ss.correlate2d(camM, Sy, mode='same')
```

```python
plt.imshow(SfilX, cmap='gray')
```

```python
plt.imshow(SfilY, cmap='gray')
```

```python
Sfil = SfilX + SfilY
plt.imshow( Sfil, cmap='gray')
```
