```python
import math
import numpy as np
import matplotlib.pyplot as plt
```

```python
genius = plt.imread('./imgs/einstein.tif')
plt.imshow(genius, cmap='gray')
```

```python
genius
```

```python
def sliceOnBit(M,b):
    h,k = M.shape
    accum = np.zeros((h,k))
    # 8 bit masks
    ms = [1,2,4,8,16,32,64,128 ]
    for i in range(h):
        for j in range(k):
            accum[i,j] = 255 if M[i,j] & ms[b-1] > 0 else 0
    
    return accum
    
```

```python
msb = sliceOnBit(genius, 7)
plt.imshow(msb, cmap='gray')
```

```python
fig, axs = plt.subplots(4, 2)
[axi.set_axis_off() for axi in axs.ravel()]
plt.tight_layout()
axs[0, 0].imshow(sliceOnBit(genius, 8), cmap='gray')
axs[0, 0].set_title('Bit plane: 8')
axs[0, 1].imshow(sliceOnBit(genius, 7), cmap='gray')
axs[0, 1].set_title('Bit plane: 7')
axs[1, 0].imshow(sliceOnBit(genius, 6), cmap='gray')
axs[1, 0].set_title('Bit plane: 6')
axs[1, 1].imshow(sliceOnBit(genius, 5), cmap='gray')
axs[1, 1].set_title('Bit plane: 5')
axs[2, 0].imshow(sliceOnBit(genius, 4), cmap='gray')
axs[2, 0].set_title('Bit plane: 4')
axs[2, 1].imshow(sliceOnBit(genius, 3), cmap='gray')
axs[2, 1].set_title('Bit plane: 3')
axs[3, 0].imshow(sliceOnBit(genius, 2), cmap='gray')
axs[3, 0].set_title('Bit plane: 2')
axs[3, 1].imshow(sliceOnBit(genius, 1), cmap='gray')
axs[3, 1].set_title('Bit plane: 1')
```
