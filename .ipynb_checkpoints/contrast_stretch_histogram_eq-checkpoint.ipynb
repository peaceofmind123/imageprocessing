```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
def equalizeHistogram(M):
    h,k = M.shape
    accum = np.zeros((h,k))
    hist = np.zeros(256)
    cumul = np.zeros(256)
    
    for i in range(h):
        for j in range(k):
            hist[M[i,j]] += 1
    ac = 0
    for i in range(256):
        ac+=hist[i]
        cumul[i] = ac
        # normalize
        cumul[i] *= 255/(h*k)
    
    for i in range(h):
        for j in range(k):
            accum[i,j] = int(cumul[M[i,j]])
    
    
    return accum

    
    
```
