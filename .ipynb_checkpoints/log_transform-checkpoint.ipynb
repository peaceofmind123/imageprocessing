```python
import numpy as np
import math
import matplotlib.pyplot as plt
```

```python
def logTransform(M,c):
    h,k = M.shape
    accum = np.zeros((h,k))
    for i in range(h):
        for j in range(k):
            accum[i,j] = c* math.log(1 + M[i,j])
    
    return accum
```

```python
marsMoon = plt.imread('./imgs/mars_moon.tif')
plt.imshow(marsMoon, cmap='gray')
```

```python
img_log = logTransform(marsMoon, 4)
plt.imshow(img_log, cmap='gray')
```

```python
dft = plt.imread('./imgs/dft.tif')
plt.imshow(dft, cmap='gray')
```

```python
dft_log = logTransform(dft, 1)
plt.imshow(dft_log, cmap='gray')
```
