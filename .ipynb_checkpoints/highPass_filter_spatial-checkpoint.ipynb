```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.signal as ss
```

```python
kernel = np.ones((3,3))*-1
kernel[1,1] = 8
kernel = kernel*1/9
```

```python
cameraMan = plt.imread('./imgs/cameraMan.tif')
plt.imshow(cameraMan, cmap='gray')
```

```python
filtered = ss.correlate2d(cameraMan, kernel, mode='same')
```

```python
plt.imshow(filtered, cmap='gray')
```

```python
hbf = 1*cameraMan + filtered
```

```python
plt.imshow(hbf, cmap='gray')
```
