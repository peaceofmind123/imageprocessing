```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
def gammaCorrect(M):
    return M**0.4
```

```python
tibia = plt.imread('./imgs/tibia.tif')
plt.imshow(tibia, cmap='gray')
```

```python
corrected = gammaCorrect(tibia)
plt.imshow(corrected, cmap='gray')
```

```python
def powerLaw(M, y):
    return M**y
```

```python
city = plt.imread('./imgs/city.tif')
plt.imshow(city, cmap='gray')
```

```python
city_p = powerLaw(city,2.9)
plt.imshow(city_p, cmap='gray')
```
