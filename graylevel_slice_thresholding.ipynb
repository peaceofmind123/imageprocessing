```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
def grayLevelSliceLinear(t1,t2, v,M):
    filter_arr = np.logical_and(M>=t1 , M <=t2)
    accum= np.copy(M)
    accum[filter_arr] = v
    
    # the rest remain untouched
    return accum
```

```python
kidney = plt.imread('./imgs/kidney.tif')
plt.imshow(kidney, cmap='gray')
```

```python
sliced = grayLevelSliceLinear(145, 255, 255, kidney)
plt.imshow(sliced, cmap='gray')
```

```python
def grayLevelSliceSlope(t1,t2,v1,v2,M):
    if t1 == t2:
        raise ValueError("Invalid threshold")
    
    m = (v2-v1)/(t2-t1)
    accum = np.copy(M)
    h,k = M.shape
    for i in range(h):
        for j in range(k):
            if accum[i,j] >=t1 and accum[i,j] <=t2:
                accum[i,j] = m*(accum[i,j] - t1) + v1
    
    return accum
```

```python
sliced_slp = grayLevelSliceSlope(150, 255, 220,255, kidney)
plt.imshow(sliced_slp, cmap='gray')
```

```python
def threshold(M,t):
    accum = np.copy(M)
    accum[accum>t] = 255
    accum[accum<=t] = 0
    return accum
```

```python
thr = threshold(kidney,130)
plt.imshow(thr, cmap='gray')
```
